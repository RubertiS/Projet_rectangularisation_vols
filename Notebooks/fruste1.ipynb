```python
# Création d'une interpolation linéaire.

N = 50  # Nombre de points de mesure sélectionnés
Y = []  # Liste pour stocker les données interpolées
Cols = []  # Liste pour stocker les noms des colonnes
for df in dsn1:
    x = df["ALTN[-]"]
    t0 = x.index[0].value
    t1 = x.index[-1].value
    t = pd.to_datetime(np.linspace(t0,t1,N))
    y = np.interp(t,x.index,x.values)
    Y.append(y)
    Cols.append(df.index.name)
ALT = pd.DataFrame(data=np.vstack(Y).T,columns=Cols)
```

4. Calcul de l'erreur de reconstruction.

```python
res = pd.DataFrame([], columns = ["NB", "SUM", "MAE", "MSE","MAX"])
for df in dsn1:
    x = df["ALTN[-]"]
    xr = x.values
    s = x.index
    t0 = s[0].value
    t1 = s[-1].value
    t = pd.to_datetime(np.linspace(t0,t1,N))
    col = df.index.name
    y = ALT[col]
    yr = np.interp(s,t,y)
    NB = len(xr)
    SUM = sum(np.abs(yr-xr))
    MAE = SUM/NB
    MSE = np.sum((yr - xr) ** 2) / NB
    MAX = np.max(np.abs(yr-xr))
    df1 = pd.DataFrame({"NB": NB, "SUM": SUM, "MAE": MAE,"MSE": MSE, "MAX": MAX}, index=[df.index.name])
        # Concaténation du DataFrame 'df1' à 'res'
    res = pd.concat([res, df1], ignore_index=True)
```

2️⃣ Filtrage des données DFDR

L’objectif est de supprimer les anomalies et le bruit, ce qui est crucial avant la compression.
🔹 Détection et suppression des valeurs aberrantes (outliers)

Méthode basée sur l'IQR (Interquartile Range).

```python
# Filtrage des outliers par l'IQR
for df in dsn1:
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    df_filtered = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

    print(f"Données après filtrage : {df_filtered.shape}")
```
