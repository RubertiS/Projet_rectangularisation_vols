```python
datadir = "C:/Users/master/Desktop/syncronisation de vols/Data/cleaned/"
final_file = datadir + "AFL1EB_cleaned_final.h5"
# Chargement des données dans un Opset (supposons que tbt.Opset est une classe personnalisée)
dsc = tbt.Opset(final_file)

# Ajout d'une référence normalisée à utiliser pour la transformation
dsn1 = tbt.Opset(norm_file).clean()  # Nettoyage des données

# Normalisation de l'altitude
for df in dsc:
    alt = df["ALT[m]"]  # Extraction de la colonne ALT[m]
    max_alt = np.max(alt)  # Calcul de l'altitude maximale
    min_alt = np.min(alt)  # Calcul de l'altitude minimale
    altn = (alt - min_alt) / (max_alt - min_alt)  # Normalisation de l'altitude
    df["ALTN[-]"] = altn  # Ajout de la colonne normalisée au DataFrame
    dsn1.put(df)  # Sauvegarde du DataFrame modifié dans dsn1

# Récupération du DataFrame final
df = dsn1.df

# Extraction des données normalisées et de leur index temporel
x = df["ALTN[-]"]  # Données normalisées
print(x.to_list())
```

```python
import pandas as pd

# Liste des clés disponibles
keys = ['/record_00', '/record_01', '/record_02', '/record_03', '/record_04', 
        '/record_05', '/record_11', '/record_12', '/record_13', '/record_14', 
        '/record_15', '/record_16', '/record_17', '/record_18', '/record_19', 
        '/record_20', '/record_21', '/record_22', '/record_23', '/record_24', 
        '/record_25', '/record_26', '/record_27', '/record_28', '/record_29', 
        '/record_30', '/record_31', '/record_32', '/record_33', '/record_34', 
        '/record_35', '/record_36', '/record_37', '/record_38', '/record_39', 
        '/record_40', '/record_41', '/record_42', '/record_43', '/record_44', 
        '/record_45', '/record_46', '/record_47', '/record_48', '/record_49', 
        '/record_50', '/record_51']

# Initialiser une liste pour stocker les valeurs d'ALT
alt_values = []

# Parcourir toutes les clés
for key in keys:
    try:
        # Charger les données pour le record courant
        data = pd.read_hdf(final_file, key=key)
        
        # Extraire la valeur de ALT (par exemple, la première ligne)
        alt_value = data.iloc[0]  # Ou data.loc['ALT'] si les lignes sont nommées
        
        # Ajouter la valeur à la liste
        alt_values.append(alt_value)
    except KeyError:
        print(f"Clé {key} non trouvée dans le fichier.")

# Créer un DataFrame avec les valeurs d'ALT
alt_df = pd.DataFrame(alt_values, index=[key.split('/')[-1] for key in keys]).T
alt_df.columns.name = 'record'
alt_df = alt_df.rename(index={0: 'ALT'}).reset_index().rename(columns={'index': 'variable'})

# Afficher le résultat
print(alt_df)
```
