```python
import pandas as pd
import numpy as np
from tabata.opset import Opset
from dtaidistance import dtw
import matplotlib.pyplot as plt
```

```python
def normalize_altitudes(ds):
    normalized_dfs = []
    
    for df in ds:
        min_val = df["ALT[m]"].min()
        max_val = df["ALT[m]"].max()
        df["ALT[m]"] = (df["ALT[m]"] - min_val) / (max_val - min_val)
            
        normalized_dfs.append(df)
    
    return normalized_dfs

input_file = "../Data/cleaned/AFL1EB_cleaned_final.h5"
ds = Opset(input_file)
normalized_dfs = normalize_altitudes(ds)
```

```python
normalized_file = "../Data/normalized/AFL1EB_normalized.h5"
normalized_ds = Opset(normalized_file)
normalized_ds.clean()

for df in normalized_dfs:
    normalized_ds.put(df)
```
